```python
#Optimal Portfolio
```

```python
number = 15000 #simulation number

np.random.seed(101) #set up the starting random point

all_weights = np.zeros((number,len(stocks.columns)))
ret_arr = np.zeros(number)
vol_arr = np.zeros(number)
sharpe_arr = np.zeros(number)


for ind in range(number):
    
    weights = np.array(np.random.random(4))
    
    weights = weights / np.sum(weights)
    
    all_weights[ind,:] = weights
    
    ret_arr[ind] = np.sum((log_ret.mean() * weights) *252)
    
    vol_arr[ind] = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov() * 252, weights)))
    
    sharpe_arr[ind] = ret_arr[ind]/vol_arr[ind]
```

```python
# Mathematical Optimization
```

```python
def get_ret_vol_sr(weights):
   
    weights = np.array(weights)
    ret = np.sum(log_ret.mean() * weights) * 252
    vol = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov() * 252, weights)))
    sr = ret/vol
    return np.array([ret,vol,sr])
```

```python
get_ret_vol_sr(weights)
```

```python
from scipy.optimize import minimize
```

```python
'''
Optimization works as a minimization function, since we actually want to maximize the Sharpe Ratio, 
we will need to turn it negative so we can minimize the negative sharpe (same as maximizing the postive sharpe)
'''
def neg_sharpe(weights):
    return  get_ret_vol_sr(weights)[2] * -1
```

```python
def check_sum(weights):
    '''
    Returns 0 if sum of weights is 1.0
    '''
    return np.sum(weights) - 1
```

```python
check_sum(weights)
```

```python
cons = ({'type':'eq','fun': check_sum})
bounds = ((0, 1), (0, 1), (0, 1), (0, 1))
init_guess = [0.25,0.25,0.25,0.25]
```

```python
opt_results = minimize(neg_sharpe,init_guess,method='SLSQP',bounds=bounds,constraints=cons)
opt_results
```

```python
opt_results.x
```

```python
get_ret_vol_sr(opt_results.x)
```

```python
# Efficient Frontier
```

```python
frontier_y = np.linspace(0,0.3,100) 
```

```python
def minimize_volatility(weights):
    return  get_ret_vol_sr(weights)[1] 
```

```python
frontier_volatility = []

for possible_return in frontier_y:
    # function for return
    cons = ({'type':'eq','fun': check_sum},
            {'type':'eq','fun': lambda w: get_ret_vol_sr(w)[0] - possible_return})
    
    result = minimize(minimize_volatility,init_guess,method='SLSQP',bounds=bounds,constraints=cons)
    
    frontier_volatility.append(result['fun'])
```
