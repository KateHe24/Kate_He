```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas_datareader
import datetime
```

```python
import pandas_datareader.data as web
```

```python
start = datetime.datetime(2021, 1, 1)
end = datetime.datetime(2021, 10, 1)
tesla = web.DataReader("TSLA", 'yahoo', start, end)
```

```python
tesla.head()
```

```python
tesla.head().shift(1)
```

```python
# Daily Percentage Change $ r_t = \frac{p_t}{p_{t-1}} -1$
```

```python
tesla['returns'] = (tesla['Close'] / tesla['Close'].shift(1) ) - 1
```

```python
tesla.head()
```

```python
tesla['returns'] = tesla['Close'].pct_change(1)
tesla.head()
```

```python
tesla.to_csv('Tesla_Stock.csv')
```

```python
Ford = web.DataReader("F", 'yahoo', start, end)
Ford.head()
```

```python
Ford['returns'] = Ford['Close'].pct_change(1)
Ford.head()
```

```python
Ford.to_csv('Ford_Stock.csv')
```

```python
General_Motors = web.DataReader("GM", 'yahoo', start, end)
General_Motors.head()
```

```python
General_Motors['returns'] = General_Motors['Close'].pct_change(1)
General_Motors.head()
```

```python
General_Motors.to_csv('GM_Stock.csv')
```
